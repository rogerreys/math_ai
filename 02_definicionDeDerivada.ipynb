## Ejemplo

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2*x

x = np.linspace(-10, 10, 100)
y = f(x)
dy = df(x)

plt.plot(x,y)
plt.plot(x,dy)
plt.plot([-10,10],[0,0], color="red")
plt.plot([0,0],[-20,100], color="red")
```

## Explicacion de grafica
La grafica de la derivada indica que:
De -10 a 0 sus valores son decrecientes
De 0 a 10 sus valores son crecientes
