# Funciones de muestreo en python

```python
import pandas as pd
import numpy as np
import random

## Leer base 


df = pd.read_csv("assts/AQI and Lat Long of Countries.csv")
df.head()
```

## Funciones de muestreo
### 1. Muestreo aleatorio simple

```python
alt_8 = df.sample(n=8)
alt_8
```

```python
# Muestre por fraccion
prop_25 = df.sample(frac=0.25)
prop_25.head()
```

### 2. Muestreo sistematico

```python
def systematic_sampling(df, step):
    index = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[index]
    return systematic_sample

systematic_sample = systematic_sampling(df ,3)
systematic_sample
```

### 3. Muestreo Estratificado

```python
df["estratificado"] = df.Country+","+df["AQI Category"]
(df["estratificado"].value_counts()/len(df)).sort_values(ascending=True)
```

```python
def data_estratificada(df, col_estra, val_estra, prop_estra, random_state=None):
    df_estra = pd.DataFrame(columns=df.columns)
    pos=-1
    for i in range(len(val_estra)):
        pos+=1
        if pos == len(val_estra)-1:
            ratio_len = len(df) - len(df_estra)
        else:
            ratio_len = int(len(df)*prop_estra[i])
        df_filtrado = df[df[col_estra] == val_estra[i]]
        df_temp =  df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

        df_estra=pd.concat([df_estra, df_temp])
    
    return df_estra
    
```

```python
val_estra = ["Viet Nam,Unhealthy for Sensitive Groups","Oman,Unhealthy","Estonia,Moderate","Jamaica,Moderate","Afghanistan,Unhealthy for Sensitive Groups", "Germany,Good","Brazil,Good","Russian Federation,Good","United States of America,Good","United States of America,Moderate"]

prop_estra = [0.000060,0.000060,0.000060,0.000060,0.000060, 0.035580, 0.039712, 0.043965, 0.084277, 0.140102]

df_estra = data_estratificada(df, "estratificado", val_estra, prop_estra, random_state=10)
df_estra



```
