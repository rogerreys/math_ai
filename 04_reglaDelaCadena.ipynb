# Composición de funciones

$$y = f(g(t))$$
$$y'(f(g(t))) = f'(g(t)) * g'(t)$$
$$\frac{dy}{dt} = \frac{dy}{dg}*\frac{dg}{dt}$$
 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    # dg/dx
    # Pendiente m=4
    # => dg/dx = 4
    return 4*x

def f(z):
    # dz/dg → entra la funcion g(x)
    # Pendiente m=1/2
    # => dz/dg = 1/2
    return (1/2)*z

def df(x):
    # f(g(x)) = df(g(x))*dg(x)  = dz
    # dz/dx = dz/dg * dg/dx
    # dz/dx =  1/2  *  4
    # dz/dx =   2
    h=0.000001
    return (f(g(x)+h)-f(g(x)))/h * (g(x+h)-g(x))/h

edad = np.linspace(0,10)
altura = g(edad)
peso = f(g(edad))
dy = df(edad)
print(dy)



plt.plot(edad,altura)
plt.plot(edad,peso)
plt.plot(edad,dy)
plt.grid()
```
