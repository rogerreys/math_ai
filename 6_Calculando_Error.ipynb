```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1.2,2,3.2,2.5,5,6,4,8], dtype='float64')
y = np.array([2,3,3.4,3.1,4,4.7,3.8,7], dtype='float64')

b=0.5
c=1
def f(x):
    return c*x+b

plt.plot(x,y, 'ro', label="dato real (y)")
x_e = np.linspace(1,8,7)
y_e = f(x_e)
plt.plot(x_e,y_e, label="prediccion (y')")
plt.legend()
plt.grid()
```

Calcular la diferencia entre el error y la estimacion

$$err=y'-y$$

Se requiere valores positivos para calcular el error 

$$|err|=|y'-y|$$

รณ tambien se puede usar la elevacion al cuadrado
$$err^2=(y'-y)^2$$

La ecuacion de error seria

$$E = \sum \limits _{i=1} ^{N} (y'-y)^2$$

EL Error Cuadratico Medio (ECM)
$$ECM = \frac{1}{N} \sum \limits _{i=1} ^{N} (y'-y)^2$$

La funcion de error o de coste nos indica que tanto nos estamos equivocando
