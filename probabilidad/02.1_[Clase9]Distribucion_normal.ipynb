```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm 
```

## Distribución normal teórica


$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$ (mu): media de la distribución
* $\sigma$ (sigma): desviación estándar de la distribución

```python
# definimos nuestra distribución gaussiana
def gaussian(x, mu, sigma):
  return (1/sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))
```

```python
x = np.arange(-4,4,0.1)
y = gaussian(x, 0.0, 1) # mu , sigma
# y = gaussian(x, 1, 0.5)

plt.plot(x, y)
```

```python
# Distribucion normal usando scipy
dis = norm(0,1)
x = np.arange(-4,4,0.1)
y = [dis.pdf(value) for value in x]
plt.plot(x,y)
```

```python
# calculando la distribución acumulada
dist = norm(0,0.5)
x = np.arange(-4,4,0.1)
y = [dist.cdf(value) for value in x]
plt.plot(x,y)
```

## Distribución normal (gausiana) a partir de los datos

* *El archivo excel* lo puedes descargar en esta página: https://seattlecentral.edu/qelp/sets/057/057.html

```python
df = pd.read_excel('s057.xls')
arr = df['Normally Distributed Housefly Wing Lengths'][3:]
val, fre = np.unique(arr, return_counts=True)
plt.bar(val, fre)
```

```python
# estimación de la distribución de probabilidad
mu = arr.mean()
des = arr.std()
# distribución teórica
x = np.arange(30,56, 0.1)
dis_nor = norm(mu, des)
y = [dis_nor.pdf(val) for val in x]
plt.plot(x,y)
# datos reales
val, freq = np.unique(arr, return_counts=True)
plt.bar(val, freq/len(arr))
```
