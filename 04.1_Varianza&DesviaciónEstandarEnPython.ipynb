```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

Leemos la base de datos a trabajar en nuestra ruta de origen

```python
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# using the attribute information as the column names
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
iris =  pd.read_csv(csv_url, names = col_names)

plt.hist(iris["Sepal_Length"], color="red", bins=int(180/5))
```

```python
sns.histplot(iris["Sepal_Length"], kde=True, bins=int(180/5), color="green")
```

Calculo de la varianza

$$\sigma^2 = \frac{\sum_{i=1}^N (x_i-\mu)^2}{N}$$

```python
iris["Sepal_Length"].var()
```

$$\sigma = \sqrt{\sigma^2} = \sqrt{\frac{\sum_{i=1}^n (x_i-\mu)^2}{N}}$$

```python
iris["Sepal_Length"].std()
```

```python
iris["Sepal_Length"].mean()
```

### Creacion de una muestra mediante el muestreo aleatorio simple

```python
muestra = iris.sample(frac=.5)
```

Calculo varianza

```python
muestra["Sepal_Length"].var()
```

Calculo de la desviacion estandar

```python
muestra["Sepal_Length"].std()
# una desviacion estandar cercano a 1 es una  desviacion estandar elevada
```

Finalmente mediab

```python
muestra["Sepal_Length"].mean()
```
